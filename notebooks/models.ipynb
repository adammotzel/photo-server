# Model Exploration

Download the model using `scripts/models/download.py`.

I manually changed the labels of all dog breeds to "dog" in `config.json`. Kinda lazy, but a simple approach to "effectively" making it a binary classifier.

```python
import torch
from PIL import Image
from transformers import AutoImageProcessor, AutoModelForImageClassification
```

```python
model_path = "../models/efficientnet-b0"

processor = AutoImageProcessor.from_pretrained(model_path)
model = AutoModelForImageClassification.from_pretrained(model_path)
```

```python
image = Image.open("../src/photos/1000002316.jpg")

inputs = processor(image, return_tensors="pt")

with torch.no_grad():
    logits = model(**inputs).logits

# model predicts one of the 1000 ImageNet classes
predicted_label = logits.argmax(-1).item()
print(model.config.id2label[predicted_label])
```
